# Match t-test

Fitness is simply how close the outputted t-stat matches the actual t-test's t-stat.

Can the GP discover the t-test with a "perfect" fitness function?

```python
!cat outputs/match-ttest-1.o
```

```python
from glob import glob

for filename in glob('outputs/match-ttest*.o'):
    with open(filename) as in_file:
        best_score = 0.
        best_ind = ''
        for line in in_file:
            line = line.strip()
            if line.startswith('1000'):
                best_score = float(line.split('\t')[4])
            elif line.startswith('0 '):
                best_ind = line
    if best_score < 1000.:
        print(best_ind)
```

Only a handful of the runs discovered the t-test, and even then they didn't consider the size of both distributions because we always simulate the same sample size.

# Different sample size

Now the normal distributions have different sample sizes, which should encourage the GP to consider the sizes of both samples.

This also resulted in removing several of the array-vs-array operations, such as x1+x2, etc.

```python
!cat outputs/evolve-stats-diff-samplesize-1.o
```

```python
from glob import glob

for filename in glob('outputs/*diff-samplesize*.o'):
    ttest_fitness = 0.
    with open(filename) as in_file:
        for line in in_file:
            line = line.strip()
            
            if 't-test fitness' in line:
                ttest_fitness = float(line.split(',')[0].split('(')[-1])
            
            if ('x1' in line and 'x2' in line and
                float(line.split(',')[0].split('(')[-1]) == ttest_fitness and
                float(line.split(',')[1]) <= 0.2):
                print(line)
        print('')
```

# Change null dist ranking to mid dist

Now the null dist ranking is according to how far away the null t-stat is from the permuted t-stats.

This change makes it so the hand-coded t-test gets 100% accuracy.

```python
!cat outputs/evolve-stats-mid-dist-1.o
```

```python
from glob import glob

for filename in glob('outputs/*mid-dist*.o'):
    if 'mid-dist-invert-invariance' in filename:
        continue
    with open(filename) as in_file:
        for line in in_file:
            line = line.strip()
            if 't-test' in line:
                print(line)
            if ('x1' in line and 'x2' in line and
                float(line.split(',')[0].split('(')[-1]) == 1. and
                float(line.split(',')[1]) <= 0.2):
                print(line)
        print('')
```

Geting really close to evolving the t-test, but not quite there. Maybe there isn't a big enough difference in the invariance score between the discovered tests and the t-test.

# Mid dist + inverted invariance

Same as mid dist but inverting the invariance score and maximizing it instead. Maybe this will help make the difference in invariance scores large enough to discover the t-test.

```python
from glob import glob

for filename in glob('outputs/*mid-dist-invert-invariance*.o'):
    with open(filename) as in_file:
        if 'nokill' in filename:
            continue
        for line in in_file:
            line = line.strip()
            if 't-test' in line:
                print(line)
            if ('x1' in line and 'x2' in line and
                float(line.split(',')[0].split('(')[-1]) == 1. and
                100. >= float(line.split(',')[1]) >= 60.):
                print(line)
        print('')
```

# Mid dist + inverted invariance + no kill

In all previous experiments, we've been "killing" individuals that don't have both `x1` and `x2` in them, as well as individuals that have a float `1.0` in them (we generally don't want constants in the solutions). Maybe we're killing off potential stepping stones for the GP. I've removed this "kill" criteria to see what happens.

```python
from glob import glob

for filename in glob('outputs/*mid-dist-invert-invariance-nokill*.o'):
    with open(filename) as in_file:
        for line in in_file:
            line = line.strip()
            if 't-test' in line:
                print(line)
            if ('x1' in line and 'x2' in line and
                float(line.split(',')[0].split('(')[-1]) == 1. and
                100. >= float(line.split(',')[1]) >= 60.):
                print(line)
        print('')
```
